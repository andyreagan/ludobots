```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# count the avg number of generations it took to get here
fitness_threshold = 1.5
threshold_avgs_easy = np.zeros([3,1])
threshold_stds_easy = np.zeros([3,1])
threshold_avgs_hard = np.zeros([3,1])
threshold_stds_hard = np.zeros([3,1])
threshold_avgs_harder = np.zeros([3,1])
threshold_stds_harder = np.zeros([3,1])
difficulty = 'easy'
for num_perturbations in range(3):
    fname = 'fitness-{0}additional-{1}.csv'.format(num_perturbations,difficulty)
    print(fname)
    # data = np.genfromtxt(fname)
    data = np.zeros([20,1000])
    f = open(fname,'r')
    for i in range(20):
        data[i,:] = np.array(list(map(float,f.readline().rstrip().split(','))))
    # print(data)
    # print(data.shape)
    generations_needed = np.zeros([data.shape[0],1])
    for i in range(data.shape[0]):
        j = 0
        fitness = data[i,j]
        while fitness<fitness_threshold:
            j+=1
            if j==1000:
                break
            else:
                fitness = data[i,j]
        generations_needed[i] = j
    # print(generations_needed)
    threshold_avgs_easy[num_perturbations] = np.mean(generations_needed)
    threshold_stds_easy[num_perturbations] = np.std(generations_needed)
difficulty = 'hard'
for num_perturbations in range(3):
    fname = 'fitness-{0}additional-{1}.csv'.format(num_perturbations,difficulty)
    print(fname)
    # data = np.genfromtxt(fname)
    data = np.zeros([20,1000])
    f = open(fname,'r')
    for i in range(20):
        data[i,:] = np.array(list(map(float,f.readline().rstrip().split(','))))
    # print(data)
    # print(data.shape)
    generations_needed = np.zeros([data.shape[0],1])
    for i in range(data.shape[0]):
        j = 0
        fitness = data[i,j]
        while fitness<fitness_threshold:
            j+=1
            if j==1000:
                break
            else:
                fitness = data[i,j]
        generations_needed[i] = j
    # print(generations_needed)
    threshold_avgs_hard[num_perturbations] = np.mean(generations_needed)
    threshold_stds_hard[num_perturbations] = np.std(generations_needed)
difficulty = 'harder'
for num_perturbations in range(3):
    fname = 'fitness-{0}additional-{1}.csv'.format(num_perturbations,difficulty)
    print(fname)
    # data = np.genfromtxt(fname)
    data = np.zeros([20,1000])
    f = open(fname,'r')
    for i in range(20):
        data[i,:] = np.array(list(map(float,f.readline().rstrip().split(','))))
    # print(data)
    # print(data.shape)
    generations_needed = np.zeros([data.shape[0],1])
    for i in range(data.shape[0]):
        j = 0
        fitness = data[i,j]
        while fitness<fitness_threshold:
            j+=1
            if j==1000:
                break
            else:
                fitness = data[i,j]
        generations_needed[i] = j
    # print(generations_needed)
    threshold_avgs_harder[num_perturbations] = np.mean(generations_needed)
    threshold_stds_harder[num_perturbations] = np.std(generations_needed)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0.2,0.2,0.7,0.7])
ax.bar(np.arange(3)+.07,threshold_avgs_easy,width=0.2,color='b') #,yerr=threshold_stds_easy)
ax.bar(np.arange(3)+.34,threshold_avgs_hard,width=0.2,color='r') #,yerr=threshold_stds_hard)
ax.bar(np.arange(3)+.61,threshold_avgs_harder,width=0.2,color='g') #,yerr=threshold_stds_hard)
ax.legend(('easy','hard','harder'),loc='best')
ax.set_xticks(np.arange(3)+0.5)
ax.set_xticklabels(('k=0','k=1','k=2'))
ax.set_xlabel('number of perturbations')
ax.set_ylabel('avg num generations for success')
plt.savefig('fitness-threshold-averages.pdf')
```
